```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
import joblib
import pickle
```

```python
def read_data(file):
    data = []
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            label = ' '.join(line[1:line.find("]")].strip().split())
            text = line[line.find("]")+1:].strip()
            data.append([label, text])
        f.close()
    return data
```

```python
def convert_label(item, name): 
    items = list(map(float, item.split()))
    label = ""
    for idx in range(len(items)): 
        if items[idx] == 1: 
            label += name[idx] + " "
    
    return label.strip()
```

```python
file = './public/data.txt'
data = read_data(file)
```

```python
emotions = ["joy", 'fear', "anger", "sadness", "disgust", "shame", "guilt"]
```

```python
tfidf = TfidfVectorizer(ngram_range = (1,2))
X = []
y = []
```

```python
for label, text in data:
    X.append(text)
    y.append(convert_label(label, emotions))
```

```python
model_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2))),
    ('classifier', LinearSVC(max_iter=100000, random_state=123))
])
```

```python
model_pipeline.fit(X, y)
```

```python
text = input("Enter Text : ")
print("Emotion : ", model_pipeline.predict([text]))
```

```python
with open('./public/lsvc_pipeline.pkl', 'wb') as f:
    pickle.dump(model_pipeline, f, protocol=4)
```
